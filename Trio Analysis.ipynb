```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('D:\\Kargo\\Risk Analysis\\ytd_trio_analysis.csv')
df[:10]
```

```python
dates = df.pivot_table('APD',index='DATE')
bom = dates[['2017-01-01','2017-02-01','2017-03-01','2017-04-01','2017-05-01','2017-06-01', \
            '2017-07-01','2017-08-01']]
bom.plot(subplots=True, figsize=(10,6), grid=False, title="APD at BOM")
```

```python
df.groupby('DOMAIN')['APD','TET'].mean().plot(kind="bar")
```

```python
def getMonth(date):
    return str(date).split('-')[1]

df['MONTH'] = df['DATE'].apply(lambda x: getMonth(x))
df.groupby('MONTH').mean()
```

```python
df.groupby('MONTH').mean().plot(grid=True, legend=False)
```
